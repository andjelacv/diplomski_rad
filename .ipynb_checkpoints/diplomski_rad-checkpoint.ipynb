<h3>Feature engineering</h3>

```python
df.info()
```

```python
#atribut duzina url predstavlja broj karaktera u URL adresi
df['duzina_url']=df['url'].apply(len)
```

```python
def duzina_TLD(url):
    extracted=tldextract.extract(url)
    return len(extracted.suffix)
```

```python
sample_url=df.url.sample().iloc[0]
TLD=(tldextract.extract(sample_url))
print('URL ADRESA:',sample_url)
print('TOP LEVEL DOMAIN:',TLD.suffix)
print('duzina_TLD:',duzina_TLD(sample_url))
```

```python
#atribut duzina_TLD predstavlja broj karaktera u najvisem nivou domenu (top level domain)
df['duzina_TLD']=df['url'].apply(duzina_TLD)
```

```python
def duzina_SLD(url):
    extracted=tldextract.extract(url)
    return len(extracted.domain)
```

```python
sample_url=df.url.sample().iloc[0]
SLD=(tldextract.extract(sample_url))
print('URL ADRESA:',sample_url)
print('SECOND LEVEL DOMAIN:', SLD.domain)
print('duzina_SLD:',duzina_SLD(sample_url))
```

```python
#atribut duzina_SLD predstavlja broj karaktera drugog nivoa domena (second level domenu)
df['duzina_SLD']=df['url'].apply(duzina_SLD)
```

```python
def duzina_subdomena(url):
    extracted=tldextract.extract(url)
    return len(extracted.subdomain)
```

```python
sample_url=df.url.sample().iloc[0]
subdomain=(tldextract.extract(sample_url))
print('URL ADRESA:',sample_url)
print('SUBDOMAIN:', subdomain.subdomain)
print('duzina_subdomena:',duzina_subdomena(sample_url))
```

```python
#atribut duzina_subdomena predstavlja broj karaktera u subdomenu(poddomenu)
df['duzina_subdomena']=df['url'].apply(duzina_subdomena)
```

```python
def ukupna_duzina_domena(url):
    extracted=tldextract.extract(url)
    if extracted.subdomain:
        full_domain=f"{extracted.subdomain}.{extracted.domain}.{extracted.suffix}"
    else:
        full_domain=f"{extracted.domain}.{extracted.suffix}"
    return len(full_domain)
```

```python
sample_url=df.url.sample().iloc[0]
domain=(tldextract.extract(sample_url))
print('URL ADRESA:',sample_url)
print('ukupna duzina_domena:',ukupna_duzina_domena(sample_url))
```

```python
#atribut ukupna duzina domena - predstavlja ukupan broj karaktera u domenu (TLD+SLD+subdomen)
df['ukupna_duzina_domena']=df['url'].apply(ukupna_duzina_domena)
```

```python
def sadrzi_ipv4(url):
    
    parsed_url = urlparse(url)
    
    
    try:
        
        ip = ipaddress.ip_address(parsed_url.hostname)
        return 1 if ip.version == 4 else 0
    except ValueError:
        
        return 0
```

```python
#atribut sadrzi_ipv4 vraa 1 ukoliko URL adresa sadri IPV4 adresu, u suprotnom vraa 1
df['sadrzi_ipv4']=df['url'].apply(sadrzi_ipv4)
```

```python
def broj_tacaka(url):
    return url.count('.')
```

```python
def broj_dir(url):
    return url.count('/')
```

```python
def broj_dir2(url):
    return url.count('//')
```

```python
def broj_at(url):
    return url.count('@')
```

```python
def broj_procenata(url):
    return url.count('%')
```

```python
def broj_dolara(url):
    return url.count('$')
```

```python
def broj_upitnika(url):
    return url.count('?')
```

```python
def broj_cifara(url):
    brojevi=0
    for i in url:
        if i.isdigit():
            brojevi=brojevi+1
    return brojevi
```

```python
def broj_slova(url):
    slova=0
    for i in url:
        if i.isalpha():
            slova=slova+1
    return slova
```

<h3> DecisionTreeClassifier </h3>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Inicijalizacija DesicionTree klasifikatora
tree=DecisionTreeClassifier()

# Treniranje modela na trening skupu
tree.fit(X_train,y_train)
```

```python
y_pred = tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
def evaluate_model(model, X_test, y_test, model_name="Model"):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    cm = confusion_matrix(y_test, y_pred)
    TN, FP, FN, TP = cm.ravel()
   
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)
    a
    print(f"Rezultati za {model_name}")
    print(f"True Negatives (TN): {TN}")
    print(f"False Positives (FP): {FP}")
    print(f"False Negatives (FN): {FN}")
    print(f"True Positives (TP): {TP}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1 Score:  {f1:.4f}")
    print(f"ROC AUC:   {roc_auc:.4f}\n")
    
    plt.figure(figsize=(6,5))
    sb.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=['Benign (0)', 'Malicious (1)'],
               yticklabels=['Benign (0)', 'Malicious (1)'])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.show()
    
    # ROC kriva
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title(f"ROC Curve - {model_name}")
    plt.show()
```

```python
evaluate_model(tree, X_test, y_test, "Decision Tree")
```

<h3> GRID SEARCH CV </h3>

```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, None],           
    'min_samples_split': [2, 5],           
    'min_samples_leaf': [1, 2],            
    'class_weight': [None, 'balanced']
}

grid = GridSearchCV(
    tree,
    param_grid=params,
    cv=5,
    scoring='recall'  
)

grid.fit(X_train, y_train)

print("Najbolji parametri:", grid.best_params_)
print("Najbolji rezultat:", grid.best_score_)
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
svm_model = make_pipeline(
    StandardScaler(),
    LinearSVC(random_state=42, max_iter=10000)
)
```

```python
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test, y_pred)
TN, FP, FN, TP = cm2.ravel()

print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")
print(f"True Positives (TP): {TP}")

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print(f"\nAccuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}  <-- maximise!")
print(f"F1 Score:  {f1:.4f}")

plt.figure(figsize=(6,5))
sb.heatmap(cm2, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Benign (0)', 'Malicious (1)'],
            yticklabels=['Benign (0)', 'Malicious (1)'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')
plt.show()

y_scores = svm_model.decision_function(X_test)
roc_auc = roc_auc_score(y_test, y_scores)
print("ROC AUC:", roc_auc)

RocCurveDisplay.from_predictions(y_test, y_scores)
plt.title("ROC Curve (Linear SVM)")
plt.show()
```
