<h3>Feature engineering</h3>

```python
df.info()
```

```python
df['duzina_url']=df['url'].apply(len)
```

```python
def duzina_domena(url):
    extracted=tldextract.extract(url)
    if extracted.subdomain:
        full_domain=f"{extracted.subdomain}.{extracted.domain}.{extracted.suffix}"
    else:
        full_domain=f"{extracted.domain}.{extracted.suffix}"
    return len(full_domain)
```

```python
df['duzina_domena']=df['url'].apply(duzina_domena)
```

```python
def duzina_TLD(url):
    extracted=tldextract.extract(url)
    return len(extracted.suffix)
```

```python
def duzina_SLD(url):
    extracted=tldextract.extract(url)
    return len(extracted.domain)
```

```python
def duzina_subdomena(url):
    extracted=tldextract.extract(url)
    return len(extracted.subdomain)
```

```python
df['duzina_subdomena']=df['url'].apply(duzina_subdomena)
```

```python
df['duzina_SLD']=df['url'].apply(duzina_SLD)
```

```python
df['duzina_TLD']=df['url'].apply(duzina_TLD)
```

```python
def duzina_subdirectory(url):
    parsed = urlparse(url)
    return len(parsed.path)
```

```python
df['duzina_subdirectory']=df['url'].apply(duzina_subdirectory)
```

```python
def sadrzi_http(url):
    parsed_url = urlparse(url)
    return 1 if parsed_url.scheme in ['http'] else 0
```

```python
# Testiranje funkcije
print(sadrzi_http("http://example.com"))
print(sadrzi_http("https://example.com")) 
print(sadrzi_http("htp://example.com"))   
print(sadrzi_http("example.com"))         
```

```python
def sadrzi_https(url):
    parsed_url = urlparse(url)
    return 1 if parsed_url.scheme in ['https'] else 0
```

```python
# Testiranje funkcije
print(sadrzi_https("http://example.com"))
print(sadrzi_https("https://example.com")) 
print(sadrzi_https("htp://example.com"))   
print(sadrzi_https("example.com"))
```

```python
def sadrzi_ipv4(url):
    
    parsed_url = urlparse(url)
    
    
    try:
        
        ip = ipaddress.ip_address(parsed_url.hostname)
        return 1 if ip.version == 4 else 0
    except ValueError:
        
        return 0
```

```python
df['sadrzi_ipv4']=df['url'].apply(sadrzi_ipv4)
```

```python
def sadrzi_ipv6(url):
    # Ekstraktujemo domen i putanju iz URL-a
    parsed_url = urlparse(url)
    
    # Pokušavamo da proverimo da li je domen IPv6
    try:
        # Ako domen sadrži IPv6 adresu, on će biti detektovan kao takav
        ip = ipaddress.ip_address(parsed_url.hostname)
        return 1 if ip.version == 6 else 0
    except ValueError:
        # Ako domen nije validan IP (npr. domen sa slovima), vraćamo 0
        return 0
```

```python
df['sadrzi_ipv6']=df['url'].apply(sadrzi_ipv6)
```

```python
def broj_tacaka(url):
    return url.count('.')
```

```python
def broj_dir(url):
    return url.count('/')
```

```python
def broj_dir2(url):
    return url.count('//')
```

```python
def broj_at(url):
    return url.count('@')
```

```python
def broj_procenata(url):
    return url.count('%')
```

```python
def broj_dolara(url):
    return url.count('$')
```

```python
def broj_upitnika(url):
    return url.count('?')
```

```python
def broj_cifara(url):
    brojevi=0
    for i in url:
        if i.isdigit():
            brojevi=brojevi+1
    return brojevi
```

```python
def broj_slova(url):
    slova=0
    for i in url:
        if i.isalpha():
            slova=slova+1
    return slova
```

<h2>Kreiranje modela</h2>

<h3> DecisionTreeClassifier </h3>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Inicijalizacija DesicionTree klasifikatora
tree=DecisionTreeClassifier()

# Treniranje modela na trening skupu
tree.fit(X_train,y_train)
```

```python
y_pred = tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

TN, FP, FN, TP = cm.ravel()

print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")  
print(f"True Positives (TP): {TP}")

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}  <-- maximise!")
print(f"F1 Score:  {f1:.4f}")

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Benign (0)', 'Malicious (1)'],
            yticklabels=['Benign (0)', 'Malicious (1)'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()

from sklearn.metrics import roc_auc_score, RocCurveDisplay

# 1. Dobavljanje verovatnoća (umesto klasa)
y_proba = tree.predict_proba(X_test)[:, 1]

# 2. Računanje ROC AUC vrednosti
roc_auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", roc_auc)

# 3. Prikaz ROC krive
RocCurveDisplay.from_estimator(tree, X_test, y_test)
```

<h3> GRID SEARCH CV </h3>

```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, None],           
    'min_samples_split': [2, 5],           
    'min_samples_leaf': [1, 2],            
    'class_weight': [None, 'balanced']
}

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    tree,
    param_grid=params,
    cv=5,
    scoring='recall'  
)

grid.fit(X_train, y_train)

print("Najbolji parametri:", grid.best_params_)
print("Najbolji rezultat:", grid.best_score_)
```

```python
# Predikcije sa najboljim modelom
best_model = grid.best_estimator_
y_pred_cv = best_model.predict(X_test)

# Prikaz metrika
print(classification_report(y_test, y_pred_cv, target_names=["Benign", "Malicious"])) 
```

```python
cm2 = confusion_matrix(y_test, y_pred_cv)
TN, FP, FN, TP = cm2.ravel()

print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")  
print(f"True Positives (TP): {TP}")

precision = precision_score(y_test, y_pred_cv)
recall = recall_score(y_test, y_pred_cv)
f1 = f1_score(y_test, y_pred_cv)

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}  <-- maximise!")
print(f"F1 Score:  {f1:.4f}")

plt.figure(figsize=(6,5))
sns.heatmap(cm2, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Benign (0)', 'Malicious (1)'],
            yticklabels=['Benign (0)', 'Malicious (1)'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()


# ROC AUC 
y_proba = best_model.predict_proba(X_test)[:, 1]

#Računanje ROC AUC vrednosti
roc_auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", roc_auc)

#Prikaz ROC krive
RocCurveDisplay.from_estimator(tree, X_test, y_test)
```
